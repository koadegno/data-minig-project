```python
import pandas as pd
```

```python
data = pd.read_csv('ar41_for_ulb_all_outliers.csv', sep=";", parse_dates=True, index_col="timestamps_UTC")
```

```python
data_outliers = data[data['reason'] == 'Outliers']
other_data = data[data['reason'] != 'Outliers']
len(data)
```

RPM

```python
df_PC1 = data_outliers[(data_outliers['RS_E_RPM_PC1'] == 0)]
df_PC1['reason'] = 'Outliers RPM PC1 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_RPM_PC1'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC1])

df_PC2 = data_outliers[(data_outliers['RS_E_RPM_PC2'] == 0)]
df_PC2['reason'] = 'Outliers RPM PC2 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_RPM_PC2'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC2])
```

Oil press

```python
df_PC1 = data_outliers[(data_outliers['RS_E_OilPress_PC1'] > 650)]
df_PC1['reason'] = 'Outliers OilPress PC1 gt 650'
data_outliers = data_outliers[(data_outliers['RS_E_OilPress_PC1'] <= 650)]
data_outliers = pd.concat([data_outliers, df_PC1])

df_PC2 = data_outliers[(data_outliers['RS_E_OilPress_PC2'] > 650)]
df_PC2['reason'] = 'Outliers OilPress PC2 gt 650'
data_outliers = data_outliers[(data_outliers['RS_E_OilPress_PC2'] <= 650)]
data_outliers = pd.concat([data_outliers, df_PC2])
```

```python
df_PC1 = data_outliers[(data_outliers['RS_E_OilPress_PC1'] == 0)]
df_PC1['reason'] = 'Outliers OilPress PC1 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_OilPress_PC1'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC1])

df_PC2 = data_outliers[(data_outliers['RS_E_OilPress_PC2'] == 0)]
df_PC2['reason'] = 'Outliers OilPress PC2 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_OilPress_PC2'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC2])
```

```python
df_PC1 = data_outliers[(data_outliers['RS_E_WatTemp_PC1'] == 0)]
df_PC1['reason'] = 'Outliers WatTemp PC1 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_WatTemp_PC1'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC1])

df_PC2 = data_outliers[(data_outliers['RS_E_WatTemp_PC2'] == 0)]
df_PC2['reason'] = 'Outliers WatTemp PC2 eq 0'
data_outliers = data_outliers[(data_outliers['RS_E_WatTemp_PC2'] != 0)]
data_outliers = pd.concat([data_outliers, df_PC2])
```

```python
data_final = pd.concat([data_outliers, other_data])
data_final['reason'] = data_final['reason'].replace('Outliers', 'Other')
data_final.to_csv('ar41_for_ulb_all_outliers.csv', sep=';')
```
