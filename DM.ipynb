```python
max_temp_air = 65*1.5
max_temp_water = 100*1.5
max_temp_oil = 115*1.5

min_temp = 0.0
```

```python
df = pd.read_csv('ar41_for_ulb.csv', sep=";")
df = df.drop_duplicates()
```

```python
df_dropped = df.dropna()
```

```python
#df_motor_stopped = df_dropped[(df_dropped['RS_E_RPM_PC1'] == 0) & (df_dropped['RS_E_RPM_PC2'] == 0)]
#df_dropped.drop(df_motor_stopped.index)
```

```python
df_motor_stopped = (df_dropped['RS_E_RPM_PC1'] == 0) & (df_dropped['RS_E_RPM_PC2'] == 0)
df_dropped = df_dropped[~df_motor_stopped] 


```

Maintenance center (found with google maps sattelite view)

```python
df_maintenance1 = df[(df["lat"] <= 51.017182) & (df["lat"] >= 51.010652) & (df["lon"] >= 3.770148) & (df["lon"] <= 3.783246)]
df_maintenance2 = df[(df["lat"] <= 50.401860) & (df["lat"] >= 50.398082) & (df["lon"] >= 4.452937) & (df["lon"] <= 4.466798)]
```

Replacing values outside of the range by -1

```python
columns = ['RS_E_InAirTemp_PC1', 'RS_E_InAirTemp_PC2', 'RS_E_WatTemp_PC1', 'RS_E_WatTemp_PC2', 'RS_T_OilTemp_PC1', 'RS_T_OilTemp_PC2']
limits = [max_temp_air, max_temp_air, max_temp_water, max_temp_water, max_temp_oil, max_temp_oil]

for col, max_temp in zip(columns, limits):
    df_dropped.loc[(df_dropped[col] < min_temp) | (df_dropped[col] > max_temp), col] = -1
```
